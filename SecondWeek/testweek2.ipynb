```python
# 1. Report how many times it is better to focus than to divide one's attention based on the average scores in the dataset. Return the resulting ratio.

avg_focus = df[df['attention'] == 'focused']['score'].mean()
avg_divide = df[df['attention'] == 'divided']['score'].mean()
    
ratio = avg_focus / avg_divide

print(f"The average of focused attention is:{avg_focus}")
print(f"The average of divided attention is:{avg_divide}")
print(f"The ratio is: {ratio.round(2)}")
```
